```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
credit = pd.read_csv('default of credit card clients.csv',header = 1)
```

```python
credit.head()
```

```python
credit.describe()
```

```python
credit.info()
```

```python
credit = credit.dropna()
```

```python
credit = credit.drop(columns='ID')
```

```python
credit
```

```python
#for simplicity make all the other options = 0
credit.loc[(credit['EDUCATION'] == 4), 'EDUCATION'] = 0
credit.loc[(credit['EDUCATION'] == 5), 'EDUCATION'] = 0
credit.loc[(credit['EDUCATION'] == 6), 'EDUCATION'] = 0
```

```python
#remove rows where the bill amount is less than 0
credit = credit[credit.BILL_AMT1 >0]
credit = credit[credit.BILL_AMT2 >0]
credit = credit[credit.BILL_AMT3 >0]
credit = credit[credit.BILL_AMT4 >0]
credit = credit[credit.BILL_AMT5 >0]
credit = credit[credit.BILL_AMT6 >0]
```

```python
header = credit.dtypes.index
print(header)
```

```python
#plot the balance limit
plt.hist(credit['LIMIT_BAL'], bins=4)
plt.show()
```

```python
#plot limit balance as line
plt.plot(credit['LIMIT_BAL'])
plt.show()
```

```python
#scatterplots: first set the two features you want to plot as variables
x = credit['PAY_0']
y = credit['PAY_2']
#now plot them
plt.scatter(x,y)
plt.show()
```

```python
#box plot: useful for finding outliers or extreme variations
A = credit['BILL_AMT1']
plt.boxplot(A,0,'gd')
plt.show()
```

```python
#correlation matrix
corrMat = credit.corr()
print(corrMat)
```

```python
#covariance guages the linear degree of change between two variables. Explains how changes in 
# 1 variable changes the second
covMat = credit.cov()
print(covMat)
```

```python


# Distribution of age, with an overlay of a density plot
age = credit['AGE']
age_dist = sns.distplot(age)
age_dist.set_title("Distribution of Passengers' Ages")

```

```python
#distribution age male vs female
fig = sns.FacetGrid(credit, hue='SEX', aspect=4)
fig.map(sns.kdeplot, 'AGE', shade=True)
oldest = credit['AGE'].max()
fig.set(xlim=(0,oldest))
fig.set(title='Distribution of Age Grouped by Gender')
fig.add_legend()
```

```python
#distribution default by age


fig = sns.FacetGrid(credit, hue='default payment next month', aspect=4)
fig.map(sns.kdeplot, 'AGE', shade=True)
oldest = credit['AGE'].max()
fig.set(xlim=(0,oldest))
fig.set(title='Distribution of Age Grouped by Default')
fig.add_legend()

```
